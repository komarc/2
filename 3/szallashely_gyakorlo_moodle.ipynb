c. Az oszlop neve legyen 'Napi ár'

```sql
select *,
    CASE
        WHEN KLIMAS='i' THEN 20000+5000
        ELSE 20000
    END as 'Napi ár'
from Szoba sz LEFT JOIN Foglalas f ON sz.SZOBA_ID=f.SZOBA_FK


EXCEPT

select *,
    CASE
        WHEN KLIMAS='i' THEN 20000+5000
        ELSE 20000
    END as 'Napi ár'
from Szoba sz JOIN Foglalas f ON sz.SZOBA_ID=f.SZOBA_FK
```
