```sql
select distinct USERNEV, count(SZALLAS_ID) over (PARTITION by USERNEV)
from Vendeg v join Foglalas f on v.USERNEV=f.UGYFEL_FK
        JOIN Szoba sz on f.SZOBA_FK=sz.SZOBA_ID
        JOIN szallashely szh on sz.SZALLAS_FK=szh.SZALLAS_ID
```

```sql
--MEGOLDS
SELECT 
    case 
        when GROUPING(v.USERNEV) = 1 then 'Vgsszeg'
        else v.usernev
    end as 'gyfl',
       COUNT(DISTINCT(szh.SZALLAS_ID)) AS 'Klnbz szllshelyek szma'
FROM Vendeg v LEFT JOIN Foglalas f ON v.USERNEV = f.UGYFEL_FK
              LEFT JOIN Szoba sz ON f.SZOBA_FK = sz.SZOBA_ID
              LEFT JOIN Szallashely szh ON sz.SZALLAS_FK = szh.SZALLAS_ID
GROUP BY ROLLUP(v.USERNEV)
```

```sql
select USERNEV, count(szh.SZALLAS_ID)
from Vendeg v join Foglalas f on v.USERNEV=f.UGYFEL_FK
        JOIN Szoba sz on f.SZOBA_FK=sz.SZOBA_ID
        JOIN szallashely szh on sz.SZALLAS_FK=szh.SZALLAS_ID
group by rollup 
```

```sql
select *, YEAR(GETDATE())-YEAR(SZUL_DAT) as 'letkor', AVG(YEAR(GETDATE())-YEAR(SZUL_DAT)) over (order by SZUL_DAT rows BETWEEN 2 preceding and 2 following ) as 'eltt eutna 2 tlag letkor'
from Vendeg
```
